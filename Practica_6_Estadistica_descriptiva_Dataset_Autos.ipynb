## Realizar un Histograma de precios de los automviles


```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate the optimal number of bins using Sturges' rule
n = len(autos.price)  # Number of data points (car prices)
num_bins = int(np.log2(n)) + 1  # Sturges' rule calculation

# Create the histogram
plt.hist(autos.price, bins=num_bins)

# Add title and labels
plt.title('Distribucin de precios de autos')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')

# Display the histogram
pyplot.show()

# --- Interpretation and Conclusions ---

# Based on the shape of the histogram, describe the distribution of car prices.
# En el histograma notamos una distribucin de precios que tiende a los 5k y se hace menos frecuente en base al aumento de precio
# Comment on the range of car prices and where the majority of prices fall; el rango decae dentro de 5k-10k y vuelve a decaer fuertemente en los 15k-20k
# Are there any outliers or unusual price points? Negativo, nada esta outofbound de manera demasiado marcada, Sin embargo el precio maximo ronda los 45k

# Relate the histogram findings to the statistics calculated earlier (min, max, mode, median, mean, std).
# Do the statistics support the observations from the histogram? True, observamos que por lo menos la min, max, moda, median, mean corresponden a los datos obtenidos

# Provide any insights or conclusions about the car market based on the price distribution.
# La mayoria de los autos se encuentran en un rango de precios considerablemente accesibles, entre 5k y 15k apartir de esa cifra su frecuencia disminuye fuertemente.
```

## Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo de vehculo (body style)


```python
# Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo de vehculo (body style)

# Crear el diagrama de caja y bigotes
plt.figure(figsize=(10, 6)) # Call plt.figure to create a new figure
sns.boxplot(x='body-style', y='price', data=autos)
# Personalizar etiquetas y ttulo
plt.xlabel('Tipo de Vehculo')
plt.ylabel('Precio del Automvil')
plt.title('Distribucin del Precio segn el Tipo de Vehculo')

# Mostrar el grfico
plt.show()
```

## Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo traccin ruedas (drive wheels)

```python
# Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo traccin ruedas (drive wheels)
# Crear el diagrama de caja y bigotes
plt.figure(figsize=(10, 6)) # Call plt.figure to create a new figure
sns.boxplot(x='drive-wheels', y='price', data=autos)
# Personalizar etiquetas y ttulo
plt.xlabel('Tipo de Traccin')
plt.ylabel('Precio del Automvil')
plt.title('Distribucin del Precio segn el Tipo de Traccin')

# Mostrar el grfico
plt.show()
```

## Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y la ubicacin del motor (engine-location)

```python
#Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y la ubicacin del motor (engine-location)
# Crear el diagrama de caja y bigotes
plt.figure(figsize=(10, 6)) # Call plt.figure to create a new figure
sns.boxplot(x='engine-location', y='price', data=autos)
# Personalizar etiquetas y ttulo
plt.xlabel('Localizacin del Motor')
plt.ylabel('Precio del Automvil')
plt.title('Distribucin del Precio segn la Localizacin del Motor')

# Mostrar el grfico
plt.show()
```
