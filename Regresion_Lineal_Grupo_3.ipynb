Modelo de Regresión Lineal


```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df1 = pd.read_csv('grupo3.csv')

X = df1[['HorasEstudio']]
Y = df1['Calificacion']

# Crear el modelo de regresión
modelo = LinearRegression()
modelo.fit(X, Y)

# Obtener pendiente e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Mostrar la ecuación de la recta
print(f"La ecuación de la línea es: y = {pendiente:.4f}x + {intercepto:.4f}")
```

```python
import pandas as pd
from scipy.stats import pearsonr


df1 = pd.read_csv('grupo3.csv')

# Extraer las variables
x = df1['HorasEstudio']           # Variable independiente: horas de estudio
y = df1['Calificacion']    # Variable dependiente: calificaciones

# Calcular el coeficiente de correlación de Pearson
coef, p_value = pearsonr(x, y)

print(f"Coeficiente de correlación de Pearson: {coef:.4f}")
print(f"Valor p: {p_value:.4f}")
```

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Cargar el archivo CSV
df = pd.read_csv('grupo3.csv')

# Renombrar columnas para facilitar el acceso
df.rename(columns={'HorasEstudio': 'estudio', 'Calificacion': 'calificaciones'}, inplace=True)

# Extraer variables
x = df['estudio']
y = df['calificaciones']

# Calcular coeficiente de correlación de Pearson
coef, p_value = pearsonr(x, y)
print(f"Coeficiente de correlación de Pearson: {coef:.4f}")
print(f"Valor p: {p_value:.4f}")

# Redimensionar para modelo de regresión lineal
X = x.values.reshape(-1, 1)
Y = y.values

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, Y)

# Predicciones
y_pred = modelo.predict(X)

# Mostrar parámetros del modelo
print(f"Pendiente (coeficiente): {modelo.coef_[0]:.4f}")
print(f"Intercepto: {modelo.intercept_:.4f}")

# Graficar datos y línea de regresión
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', linewidth=2, label='Línea de regresión')
plt.title('Regresión Lineal: Horas de Estudio vs Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Pendiente (coeficiente): 5.2857

Intercepto (ordenada al origen): 0.1081

Interpretación:
La línea de regresión tiene una pendiente altamente positiva, lo que indica una tendencia ascendente muy fuerte: En este caso, por cada hora extra que un estudiante dedica al estudio, su calificación final aumenta en promedio 5.29 puntos. Esto muestra que el tiempo de estudio tiene un impacto fuerte y positivo sobre el rendimiento académico. Una pendiente tan alta sugiere que incluso pequeñas diferencias en el tiempo de estudio pueden producir mejoras significativas en las calificaciones. El intercepto representa el valor estimado de la calificación cuando el tiempo de estudio es cero. Es decir, si un estudiante no estudia nada (0 horas), su calificación esperada sería aproximadamente 0.1081 puntos.
