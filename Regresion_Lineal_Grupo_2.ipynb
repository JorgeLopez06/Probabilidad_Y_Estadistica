Modelo de Regresión Lineal


```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df1 = pd.read_csv('grupo2.csv')

X = df1[['HorasEstudio']]
Y = df1['Calificacion']

# Crear el modelo de regresión
modelo = LinearRegression()
modelo.fit(X, Y)

# Obtener pendiente e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Mostrar la ecuación de la recta
print(f"La ecuación de la línea es: y = {pendiente:.4f}x + {intercepto:.4f}")
```

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Cargar el archivo CSV
df = pd.read_csv('grupo2.csv')

# Renombrar columnas para facilitar el acceso
df.rename(columns={'HorasEstudio': 'estudio', 'Calificacion': 'calificaciones'}, inplace=True)

# Extraer variables
x = df['estudio']
y = df['calificaciones']

# Calcular coeficiente de correlación de Pearson
coef, p_value = pearsonr(x, y)
print(f"Coeficiente de correlación de Pearson: {coef:.4f}")
print(f"Valor p: {p_value:.4f}")

# Redimensionar para modelo de regresión lineal
X = x.values.reshape(-1, 1)
Y = y.values

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, Y)

# Predicciones
y_pred = modelo.predict(X)

# Mostrar parámetros del modelo
print(f"Pendiente (coeficiente): {modelo.coef_[0]:.4f}")
print(f"Intercepto: {modelo.intercept_:.4f}")

# Graficar datos y línea de regresión
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', linewidth=2, label='Línea de regresión')
plt.title('Regresión Lineal: Horas de Estudio vs Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Pendiente (coeficiente): 4.2227

Intercepto (ordenada al origen): 15.3075

Interpretación:
La línea de regresión tiene una pendiente altamente positiva, lo que indica una tendencia ascendente muy fuerte: a medida que aumentan las horas de estudio, las calificaciones tienden a aumentar mayusulamente. Esto concuerda con el coeficiente de correlación muy elevado, que sugiere que los datos de horas estudiadas son determinantes de las calificaciones en esta muestra de alumnos.
