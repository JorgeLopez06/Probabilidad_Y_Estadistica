Modelo de Regresin Lineal


```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df1 = pd.read_csv('grupo1.csv')

X = df1[['HorasEstudio']]
Y = df1['Calificacion']

# Crear el modelo de regresin
modelo = LinearRegression()
modelo.fit(X, Y)

# Obtener pendiente e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Mostrar la ecuacin de la recta
print(f"La ecuacin de la lnea es: y = {pendiente:.4f}x + {intercepto:.4f}")
```

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Cargar el archivo CSV
df = pd.read_csv('grupo1.csv')

# Renombrar columnas para facilitar el acceso
df.rename(columns={'HorasEstudio': 'estudio', 'Calificacion': 'calificaciones'}, inplace=True)

# Extraer variables
x = df['estudio']
y = df['calificaciones']

# Calcular coeficiente de correlacin de Pearson
coef, p_value = pearsonr(x, y)
print(f"Coeficiente de correlacin de Pearson: {coef:.4f}")
print(f"Valor p: {p_value:.4f}")

# Redimensionar para modelo de regresin lineal
X = x.values.reshape(-1, 1)
Y = y.values

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, Y)

# Predicciones
y_pred = modelo.predict(X)

# Mostrar parmetros del modelo
print(f"Pendiente (coeficiente): {modelo.coef_[0]:.4f}")
print(f"Intercepto: {modelo.intercept_:.4f}")

# Graficar datos y lnea de regresin
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', linewidth=2, label='Lnea de regresin')
plt.title('Regresin Lineal: Horas de Estudio vs Calificacin')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacin')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
