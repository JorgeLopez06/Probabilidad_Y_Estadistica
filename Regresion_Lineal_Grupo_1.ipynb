Modelo de Regresin Lineal


```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df1 = pd.read_csv('grupo1.csv')

X = df1[['HorasEstudio']]
Y = df1['Calificacion']

# Crear el modelo de regresin
modelo = LinearRegression()
modelo.fit(X, Y)

# Obtener pendiente e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Mostrar la ecuacin de la recta
print(f"La ecuacin de la lnea es: y = {pendiente:.4f}x + {intercepto:.4f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
df1 = pd.read_csv('grupo1.csv')
X = df1[['HorasEstudio']]
Y = df1['Calificacion']
sns.set(style="whitegrid")
sns.regplot(x=X, y=Y, scatter=True, color='blue')
plt.title('Relacin entre Horas de Estudio y Calificacin')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacin')
plt.grid(True)
plt.show()
```

```python
#Regeresion lineal simple
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
lm = LinearRegression()
#variables independientes (predictoras)
x = grupo1[['HorasEstudio']]
#variable dependiente (objetivo)
y = grupo1['Calificacion']
x.head(10)
```

```python
#entrenamos el algoritmo de regresion lineal
lm.fit(x,y)
```

```python
#Realizamos la fase de prediccin
Ypredict = lm.predict(x)
```

```python
#valor del coeficiente o peso de la variable

lm.coef_
```

```python
#Precision del modelo
lm.score(x,y)
```
