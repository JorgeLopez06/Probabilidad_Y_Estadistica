### Estadistica descriptiva.
#### Es la que se usa para descrbir o resumir los valores y las observaciones de un conjunto de datos, por ejemplo una aplicación de seguimiento del estado físico registró los pasos diarios de una persona y su frecuencia cadríaca durante un período de diez días.

```python
import random
edades = [] #Creamos una lista vacia
for i in range(100): #nos dara 100 numeros aleatorios
  numero_aleatorio = random.randint(18, 70)#valores entre 18 y 70
  edades.append(numero_aleatorio)#agregamos el numero aleatorio a la lista de valores
print(edades)
```

```python
import pandas as pd
```

```python
#creamos un dataframe de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns = ['Edad'])
df.head()
```

```python
#calcular la tabla de distribución de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by = 'Edad')

#imprimir la tabla de distribución de frecuencias
print("Tabla de Distribución de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia Relativa'] = tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia Porcentual'] = tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

## REGLA DE STURGES
1. Es una regla que sirve para calcular el numer ode clases o intervalos idoneo en los que se debe dividir un conjunto de datos.
2. Establece que el numero de clases es igual a uno mas el logaritmo en base dos del numero total de datos.

' c = 1 + log2(N) '



```python
#importamos la libreria math de funciones matemáticas
import math
```

```python
#llamamos de vuelta la lista edades
print(edades)
```

```python
#creamos un DataFrame con las edades
df = pd.DataFrame({'Edad':edades})
df
```

```python
#Calcular el numero de intervalos utilizando la regla de Sturges
n=len(edades)
K=int(1+math.log2(n))
print("Valor de n = ", n , " \nNúmero de intervalos = ", K)
```

```python
#Crear la tabla de frecuencias
tabla_frecuencia = pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frecuencia']
tabla_frecuencia
```

```python
#Podemos ordenar la tabla de frecuencia por el intervalos
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
#podemos generar las columnas con los nuevos rangos
tabla_frecuencia['Frecuencia Relativa']=tabla_frecuencia['Frecuencia']/n
tabla_frecuencia['Frecuencia Porcional']=tabla_frecuencia['Frecuencia']/n*100
tabla_frecuencia['Frecuencia Accumulativa']=tabla_frecuencia['Frecuencia'].cumsum
tabla_frecuencia['Frecuencia Relativa Acumulativa']=tabla_frecuencia['Frecuencia Relativa'].cumsum()
tabla_frecuencia['Frecuencia Porcional Acumulativa']=tabla_frecuencia['Frecuencia Porcional'].cumsum()

#mostrar la tabla de frecuencia
print(tabla_frecuencia)
```
