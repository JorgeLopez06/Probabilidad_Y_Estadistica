Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
Cuantos estudiantes pesan más de 68 y menos de 72 kg.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70
desv_std = 4
num_datos = 500
```

```python
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] # mostrar los primeros 50
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.6, color='b')
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(media - 3*desv_std, media + 3*desv_std, num_datos)
pdf = 1/(desv_std * np.sqrt(2 * np.pi)) * np.exp( - (x - media)**2 / (2 * desv_std**2) )
```

La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.

```python
media = 70
desv_std = 4
num_datos = 500
peso_limite = 72
prob_mas_68kg= norm.cdf(68, media, desv_std)
prob_menos_72kg= norm.cdf(72, media, desv_std)  # P(X ≤ 72)
prob_entre_68_y_72 = prob_menos_72kg - prob_mas_68kg
print(f"Probabilidad entre 68 kg y 72 kg: {prob_entre_68_y_72:.4f}")
```

Histogramas distribución de estudiantes

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros del problema
media = 70                 # Media (μ)
desv_std = 4               # Desviación estándar (σ)
total_estudiantes = 500   # Total de estudiantes

# Intervalo de interés
peso_min = 68
peso_max = 72

# Calcular los valores de la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Calcular la probabilidad acumulada entre 68 y 72 kg
prob_min = norm.cdf(peso_min, media, desv_std)
prob_max = norm.cdf(peso_max, media, desv_std)
prob_entre = prob_max - prob_min

# Número estimado de estudiantes en ese rango
estudiantes_entre = prob_entre * total_estudiantes

# Gráfica de la distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_min) & (x <= peso_max), color='skyblue', alpha=0.5)

# Título con parámetros
plt.title('Probabilidad de que un estudiante pese entre 68 kg y 72 kg,\n'
          r'$\mu=70$, $\sigma=4$, $X=500$')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()

# Resultados en consola
print(f"Probabilidad de que un estudiante pese entre {peso_min} kg y {peso_max} kg: {prob_entre:.4f}")
print(f"Número de estudiantes que pesan entre {peso_min} kg y {peso_max} kg: {estudiantes_entre:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros del problema
media = 70
desv_std = 4
total_estudiantes = 500

# Intervalo de interés
peso_min = 68
peso_max = 72

# Calcular la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

# Calcular la probabilidad de estar entre 68 y 72 kg
prob_min = norm.cdf(peso_min, media, desv_std)
prob_max = norm.cdf(peso_max, media, desv_std)
prob_entre = prob_max - prob_min
estudiantes_entre = prob_entre * total_estudiantes

# Imprimir los resultados
print(f"Probabilidad de que un estudiante pese entre 68 kg y 72 kg: {prob_entre:.4f}")
print(f"Número estimado de estudiantes entre 68 y 72 kg: {estudiantes_entre:.2f}")

# Crear la gráfica
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_min) & (x <= peso_max), color='skyblue', alpha=0.5)
plt.axvline(x=peso_min, color='red', linestyle='--', label=f'{peso_min} kg')
plt.axvline(x=peso_max, color='green', linestyle='--', label=f'{peso_max} kg')

# Personalización
plt.title('Distribución Normal del Peso de Estudiantes\n'
          r'$P(68 \leq X \leq 72)$, $\mu=70$, $\sigma=4$, $n=500$')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
