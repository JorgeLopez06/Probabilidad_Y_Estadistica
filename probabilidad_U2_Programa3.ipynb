Ejercicio permutaciones.

1. Existen 10 asistentes de profesor disponibles para calificar examenes en un curso de calculo en una gran universidad.
2. El primer examen se compone de 4 preguntas y el profesor desea seleccionar un asistente diferente para calificar cada pregunta (solo un asistente por pregunta).
3. De cuantas maneras se pueden elegir los asistentes para calificar?
4. En este caso n = tamao del grupo = 10 y k = tamao del subconjunto = 4
5. El numero de permutaciones es?

```python
import math as mt
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes = ["Aron","Brandon","Cael","Diego","Eros","Fabio","Guezo","Pablo","Victor","Dereck"]
n = len(asistentes)
r = 4

Pr = mt.factorial(n)/(mt.factorial(n-r))

print("Existen ", Pr , " posibles permutaciones sin repeticion del total de ", n, " asistentes ", " en grupos de ", r ," en ", r)

print("Aqui las permutaciones")
permutaciones_asistentes = list(permutations(asistentes,r))

for permutacion in permutaciones_asistentes:
  print(permutacion)
```
