```python
import math as mt
from itertools import combinations
from itertools import combinations_with_replacement
import numpy as np
```

Ejemplo 1. Nombres de personas

```python
nombres = ["Oscar","Paco","Paty","Laura","Rubn","Luis","Lucy","Alberto","Juan","Danny"]
n = len(nombres)
r = 2 #

Cn=mt.factorial(n)/(mt.factorial(r)*mt.factorial(n-r))

print("Existen ", Cn , " posibles combinaciones sin repeticin del total de ", n , " nombres ", " en grupos de ", r ," en ",r)

print("Aqu las combinaciones")
combinaciones_nombres=list(combinations(nombres,r))

#imprimir las combinaciones
for combinacion in combinaciones_nombres:
  print(combinacion)
```

Ejemplo 2. Caracteres alfabticos y dgitos

```python
caracteres = ["A","B","C","D","E","F","0","1","2","3","4","5"]
n = len(caracteres)
r = 10 # como agrupar

Cn = mt.factorial(n)/(mt.factorial(r)*mt.factorial(n-r))

print("Existen ", Cn , " posibles combinaciones sin repeticin del total de ", n , " caracteres ", " en grupos de ", r ," en ",r)

print("Aqu las combinaciones")
combinaciones_caracteres=list(combinations(caracteres,r))

#imprimir las combinaciones
for combinacion in combinaciones_caracteres:
  print(combinacion)
```

Ejemplo 3. Combinaciones con repeticin
