# Ejemplo 1: Registro de calificaciones de 12 alumnos

## 1. En este ejemplo generamos una lista que contiene las calificaciones de una grupo de 12 alumnos
## 2. Con esta lista de valores se pretende determinar las **Medidas de tendencia, central, es decir, la media la mediana y la moda.**


```python
# Lista de calificacioens del grupo a de probabilidadt y estradistica

grupoa = [70,70,65,90,100,95,45,80,73,20,100,96]
print(grupoa)
```

```python
# Linea indicativa del tipo de dato que otorgamos a nuestra variuable grupo a

type(grupoa)
```

## Obtener el promedio del grupo a

```python
# Sacamos la media (promedio) de las calificaciones con su formula

promedio = sum(grupoa)/len(grupoa)
print(promedio)
```

```python
# Obtenemos el mismo resultado con nuestra libreria statistic usando la funcion mean para obtener el promedio

promedio2 = mean(grupoa)
print(promedio2)
```

## Obtener la mediana del grupo a

```python
# La mediana es el valor de la lista que se encuentra en el medio, en caso de que fuera una lista de valores par, es el promedio de los dos valores centrales
# En este caso son 73 y 80 cuyo promedio es 76.5
mediana = median(grupoa)
print(mediana)
```

## Obtener la moda del grupo a

```python
# La moda es aquel valor que se repite en mayor ocasion dentro de la lista
# Podemos determinarlo con la funcion mode sin embargo son 2 modas 70 y 100 que se repiten ambos 2 veces

moda=mode(grupoa)
print(moda)
```

```python
# Usando la funcion multimode solucionamos este problema
# De esta forma multiples modas pueden existir

moda2=multimode(grupoa)
print(moda2)
```

# Ejemplo 2. Tiempos de 100 metros de un grupo de 10 alumnos
## 1. En este ejemplo generamos una lista de un grupo de 10 alumnos donde se dan los tiempos en segundos que les toma recorrer 100 metros.
## 2. Obtener media, mediana y moda.

```python
# Definimos la variable llamada tiempos100mts para la distancia de 100 metros que recorrieron los alumnos y capturamos sus tiempos
tiempos100mts = [15.10,17.20,14.69,13.27,22.15,18.71,19.15,20.65,15.10,17.20]

# Desplegamos la lista de valores
print(tiempos100mts)
```

## Obtener promedio

```python
promedio = mean(tiempos100mts)
print("El tiempo promedio de los corredores del grupo a es de =>",mean(tiempos100mts))
```

## Obtener mediana

```python
mediana = median(tiempos100mts)
print("La media de los tiempos de los corredores del grupo a es de =>",median(tiempos100mts))
```

## Obtener moda

```python
moda = multimode(tiempos100mts)
print("La moda de los tiempos de los corredores del grupo a es de =>",multimode(tiempos100mts))
```

# Ejemplo 3: Histogramas de 2 grupos


## 1. En este ejemplo generamos 100 numeros aleatorios reales que representan las producción diaria de 100 telares (medidas en metros).
## 2. Generar otra lista con la producción de 100 telares para una segunda
## 3.
## 4.

```python
# Libreria necesaria para generar numeros aleatorios
import random
```

```python
fabrica1=[]
for i in range(100):
  numero_aleatorio = round( random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)

print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aleatorio = round( random.uniform(50,100),2)
  fabrica2.append(numero_aleatorio)

print(fabrica2)
```

```python
# Libreria para graficar valores
import matplotlib.pyplot as plt
from matplotlib.text import Text
```

```python
plt.hist(fabrica1)
plt.title("Producción diaria de la Fabrica 1")

Text(0.5, 1.0, 'Producción diaria de la Fabrica 1')
```

```python
plt.hist(fabrica2)
plt.title("Producción diaria de la Fabrica 2")

Text(0.5, 1.0, 'Producción diaria de la Fabrica 2')
```

## Datos de Histogramas

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Valores de la fabrica 1")
print("Media =>",media1)
print("Mediana =>",mediana1)
print("Moda =>",moda1)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = multimode(fabrica2)

print("Valores de la fabrica 2")
print("Media =>",media2)
print("Mediana =>",mediana2)
print("Moda =>",moda2)
```

## Comparación de Histogramas

```python
# genera 2 sublplot de 1 renglon por 2 columnas para desplegar los 2 histogramas uno al lado del otro
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la Fabrica 1")
plt.legend()
plt.show()
```

```python
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label='Mediana')
for mode_value in moda2:
    plt.axvline(mode_value, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la Fabrica 2")
plt.legend()
plt.show()
```
